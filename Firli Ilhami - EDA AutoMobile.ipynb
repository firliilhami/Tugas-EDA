```python
numerik.corr(method = 'spearman').style.background_gradient().set_precision(2)
```

```python
#create funtion to investigate the relationship of each feature to the label
def scatterplot(data, features, target):
    features=list(numerik.columns)
    #loop though all features
    for index, feature in enumerate(features):
        
        #create joint plot
        sns.regplot(x = feature, y = target, data = data)
        
        #show the figure
        plt.show()
        
```
